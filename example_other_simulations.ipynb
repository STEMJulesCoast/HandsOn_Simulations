### Beispiel statische Simulation

Eine statische Simulation untersucht das Verhalten oder die Reaktion eines Systems unter festgelegten Bedingungen, ohne dass zeitliche Veränderungen oder dynamische Effekte berücksichtigt werden.
In der folgenden Code-Zelle wird der hydrostatische Druck auf einen Körper in 2 m Tiefe simuliert.

```python
def calculate_hydrostatic_pressure(rho, g, h):
    """
    Calculates the hydrostatic pressure on the wall of a container at a certain depth.

    Parameters:
    - rho: Density of the fluid (kg/m^3)
    - g: Acceleration due to gravity (m/s^2)
    - h: Depth of the fluid (m)

    Returns:
    - Pressure (in Bar)
    """
    pressure_pascal = rho * g * h  # Calculate pressure in Pascal
    pressure_bar = pressure_pascal / 100000  # Convert Pascal to Bar
    return pressure_bar

# Parameters for water
rho = 1000  # Density of water in kg/m^3
g = 9.81  # Acceleration due to gravity in m/s^2

# Depth of the water in the container
h = 2  # in meters

# Calculate the pressure
pressure = calculate_hydrostatic_pressure(rho, g, h)
print(f"The hydrostatic pressure at a depth of {h} meters is {pressure} bar.")
```

### Diskrete Simulation

Ein typisches Beispiel für eine diskrete Simulation ist die Modellierung eines Warteschlangensystems, wie es beispielsweise an Supermarktkassen  oder in der Logistik vorkommt. Hier ein einfaches Beispiel mit einer Warteschlange, in der Kunden ankommen und bedient werden.

```python
import random
import queue
import matplotlib.pyplot as plt

# Settings
simulation_time = 100  # Simulation time in time units (seconds or minutes for example)
arrival_rate = 0.1  # Average rate of arrival
service_rate = 0.2  # Average service rate

# Initialize queue
queue = queue.Queue()
arrival_times = []
service_times = []

# Simulation
current_time = 0
while current_time < simulation_time:
    if random.random() < arrival_rate:
        queue.put(current_time)
        arrival_times.append(current_time)
    
    if not queue.empty() and random.random() < service_rate:
        start_time = queue.get()
        service_times.append(current_time - start_time)
    
    current_time += 1

# Plot results
plt.hist(service_times, bins=10, alpha=0.7, label='Service Times')
plt.xlabel('Time')
plt.ylabel('Number of Customers')
plt.title('Distribution of Service Times')
plt.legend()
plt.show()
```

Die Abbildung zeigt die Verteilung der Bedienungszeiten der Kund:innen. Jede Person, der in die Warteschlange kommt, wird nach einer gewissen Zeit bedient. Diese Bedienungszeit ist die Differenz zwischen der Zeit, zu der die Person in der Warteschlange ankommt, und der Zeit, zu der die Person bedient wird.

Das Histogramm visualisiert, wie oft bestimmte Bedienungszeiten vorkommen. Auf der x-Achse sind die Bedienungszeiten abgetragen, und auf der y-Achse ist die Anzahl der Personen, die jeweils diese Bedienungszeit hatten.

Wenn es zum Beispiel viele Personen mit kurzen Bedienungszeiten gibt, zeigt das Histogramm einen hohen Balken im Bereich der kürzeren Zeiten. Sind die Bedienungszeiten dagegen breit gestreut oder gibt es viele Personen mit längeren Wartezeiten, zeigt sich das in einem breiteren Histogramm.

Das Ziel solcher Analysen kann es sein, Engpässe zu identifizieren, die Effizienz des Bedienungsprozesses zu bewerten und Verbesserungen vorzuschlagen, um die Wartezeiten zu minimieren und die Anzahl der Personen pro Kasse zu erhöhen.

### Monte-Carlo-Simulation

Die Monte-Carlo-Simulation wird verwendet, um die Wahrscheinlichkeit von verschiedenen Szenarien in unsicheren Situationen zu schätzen und wird u.A. in der Finanzwelt genutzt. Hier ein Beispiel, das die mögliche Endsumme von 1000 Investitionen mit einer angenommenen jährlichen Rendite von 5% ± 4% simuliert.

```python
import numpy as np


# Parameters
investments = 1000
years = 30
mean_return = 0.05
risk = 0.04

# Simulation
final_values = []
for _ in range(investments):
    annual_returns = np.random.normal(mean_return, risk, years)
    final_value = np.prod(1 + annual_returns)
    final_values.append(final_value)

# Plot results
plt.hist(final_values, bins=50, alpha=0.75)
plt.xlabel('Final Value')
plt.ylabel('Frequency')
plt.title('Distribution of Final Values after 30 Years')
plt.show()
```

Das Histogramm zeigt uns die Verteilung der Endwerte von 1000 simulierten Investitionen über einen Zeitraum von 30 Jahren. Dabei wird eine angenommene durchschnittliche jährliche Rendite mit einer bestimmten Varianz (Risiko) verwendet. Diese Simulation hilft dabei, das potenzielle Risiko und die möglichen Erträge einer Investition unter Berücksichtigung der Unsicherheit zu untersuchen.

Auf der x-Achse des Histogramms sind die möglichen Endwerte der Investitionen dargestellt, während die y-Achse angibt, wie viele der simulierten Investitionen zu diesen Endwerten geführt haben. 

    Breite Verteilung: Eine breite Verteilung der Endwerte deutet auf ein hohes Risiko hin, da die Investitionen zu sehr unterschiedlichen Ergebnissen führen können.
    Schmale Verteilung: Eine schmalere Verteilung zeigt ein geringeres Risiko an, da die meisten Investitionen ähnliche Endwerte erzielen.
    Links- oder Rechtsschiefe: Eine schiefe Verteilung kann darauf hinweisen, dass es wahrscheinlicher ist, höhere oder niedrigere Renditen als den Mittelwert zu erzielen.


Das Monte-Carlo-Histogramm hilft, die Unsicherheit und das Risiko von Investitionen zu visualisieren und zu verstehen.
