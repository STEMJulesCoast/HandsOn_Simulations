```python
import time
import tkinter

class VirtualPet:
    def __init__(self, name):
        self.name = name
        self.hunger = 0
        self.thirst = 0
        self.tiredness = 0
        self.boredom = 0

    def feed(self):
        self.hunger = max(0, self.hunger - 10)
        print(f"{self.name} wurde gefüttert.")

    def give_water(self):
        self.thirst = max(0, self.thirst - 10)
        print(f"{self.name} hat jetzt weniger Durst.")

    def put_to_sleep(self):
        self.tiredness = max(0, self.tiredness - 10)
        print(f"{self.name} schläft jetzt.")

    def play(self):
        self.boredom = max(0, self.boredom - 10)
        print(f"{self.name} spielt und ist weniger gelangweilt.")

    def pass_time(self):
        self.hunger += 2
        self.thirst += 2
        self.tiredness += 2
        self.boredom += 2

    def status(self):
        print(f"Status von {self.name}: Hunger {self.hunger}, Durst {self.thirst}, Müdigkeit {self.tiredness}, Langeweile {self.boredom}")

def main():
    pet_name = input("Wie möchtest du dein virtuelles Haustier nennen? ")
    pet = VirtualPet(pet_name)

    choices = {
        "1": pet.feed,
        "2": pet.give_water,
        "3": pet.put_to_sleep,
        "4": pet.play,
    }

    while True:
        print(f"\nWas möchtest du mit {pet_name} machen?")
        print("1. Füttern")
        print("2. Wasser geben")
        print("3. Schlafen legen")
        print("4. Spielen")
        print("5. Status anzeigen")
        print("6. Beenden")

        choice = input("> ")
        if choice in choices:
            choices[choice]()
        elif choice == "5":
            pet.status()
        elif choice == "6":
            print("Danke fürs Spielen!")
            break
        else:
            print("Bitte wähle eine gültige Option.")

        pet.pass_time()
        time.sleep(1.5)  # Kurze Pause, um die Interaktion realistischer zu gestalten

if __name__ == "__main__":
    main()




```
